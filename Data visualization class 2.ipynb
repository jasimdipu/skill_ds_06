```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

### relational plot

```python
sns.relplot(x='total_bill', y='tip', data=df)
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex', style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex', style='smoker', size='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex', style='smoker', size='size', col='time')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex', col='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='smoker', col='sex')
```

```python
sns.relplot(x='total_bill', y='tip', data=df, hue='sex', col='day')
```

### scatter plot

```python
sns.scatterplot(x='day', y='tip', data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

### Categorical plot

```python
sns.catplot(x='day', y='tip', data=df)
```

```python
sns.catplot(x='day', y='tip', data=df, kind='swarm')
```

```python
sns.catplot(x='day', y='tip', data=df, kind='swarm', hue='size')
```

```python
sns.catplot(x='day', y='tip', data=df, kind='box', hue='size')
```

```python
sns.catplot(x='day', y='tip', data=df, kind='box')
```

```python
sns.pairplot(data=df)
```
