# From Function

```python
def function_1(i, j):
    return i*j
```

```python
function_1(2, 3)
```

```python
func = function_1
```

```python
np.fromfunction(func, (5, 5))
```

```python
np.fromfunction(lambda i, j: i*j, (4, 4))
```

```python
np.fromfunction(lambda i, j: i==j, (4, 4))
```

```python
t = tuple(x for x in range(10))
t
```

```python
arr = np.fromiter(t, float)
arr
```

```python
arr = np.fromiter(t, int)
arr
```

```python
arr = np.fromstring("10,20,30,40,50,60,70,80,90", sep=",", dtype=int)
arr
```

```python
arr = np.fromstring("10-20-30-40-50-60-70-80-90", sep="-", dtype=int)
arr
```
